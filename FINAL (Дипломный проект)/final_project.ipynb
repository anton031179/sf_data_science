#### 2. Подготовка к сдаче итогового проекта

На работу над итоговым проектом отводится 4 недели. В течение этого срока вам необходимо будет определиться с темой, реализовать проект и загрузить решение на платформу.

#### ВЫБОР ТЕМЫ

Мы решили не ограничивать вас в выборе темы, однако понимаем, что выбор может оказаться непростым, поэтому составили небольшой список рекомендаций.

#### РЕКОМЕНДАЦИИ ПО ВЫБОРУ ИТОГОВОГО ПРОЕКТА

Мы считаем, что итоговый проект должен стать украшением вашего резюме, поэтому выберите такую тему, которая:

- будет отвечать вашим карьерным устремлениям. Об этом проекте вы, скорее всего, захотите рассказать на собеседовании или применить его результаты в текущей профессиональной деятельности.
- будет обладать определённой ценностью для бизнеса.
- продемонстрирует весь спектр навыков, которым вы научились в течение этого года. Покажите, какими инструментами вы владеете.
Не рекомендуем брать слишком сложную тему.

#### СВОЯ ТЕМА

Возможно, у вас уже есть на примете задача, над которой вы хотите детально поработать и которая отвечает приведённым выше рекомендациям. Дерзайте! Совместно с ментором подумайте, как упаковать вашу идею в проект, который покажет все грани вашего профессионализма и заинтересует ваших будущих работодателей.

#### КРИТЕРИИ ОЦЕНИВАНИЯ ПРОЕКТА

**1. Анализ и обработка данных**

**Обработка и очистка данных**  
0	Сбор данных (если датасет не выдавался заранее), очистка и EDA не проведены. Студент использует готовый датасет, не уделив внимания его предобработке.    
1	Студент произвёл формальный EDA, но результатами не пользуется, очистка не произведена (если необходима).  
2	Студент самостоятельно подготовил датасет и произвёл его качественный EDA.  
**Статистический анализ данных**    
0	Студент приводит некоторые выводы на основе данных, однако они не подкреплены статистическими тестами.  
1	Студент использует некоторые статистические тесты, но выводы не используются в дальнейшем.  
2	Студент корректно подтверждает свои выводы статистическими тестами и активно использует эту информацию в решении.  
**Использование инструментов визуализации**  
0	Студент не пользуется инструментами визуализации.  
1	Студент использует инструменты визуализации, но формально, не сделав из этого выводов.  
2	Студент активно использует визуализацию и выводы из неё в своём решении.  

**2. Применение ML и DL**

**Наличие baseline, проверка способов решения задачи**  
0	Студент не проверил, можно ли решить задачу простейшими методами, и неочевидно, что задача решается ими хуже, чем с использованием продвинутых методов ML.  
1	Студент не проверил способы решения задачи, но очевидно, что данная задача не решается наивными методами.  
2	Студент проанализировал способы решения задачи / подготовил baseline для своего решения и показывает превосходство своего решения.  
**Соответствие выбранных методов требованиям задачи**   
0	Методы кардинально не соответствуют решаемой задаче, при этом показывают низкое качество.  
1	Методы в целом соответствуют решаемой задаче.  
2	Методы соответствуют решаемой задаче / методы не соответствуют решаемой задаче, но их использование позволяет добиться существенного улучшения качества.  
**Комплексность методов (применение различных моделей, энкодеров, ансамблей и stacking и др.)**  
0	Студент необоснованно не использует комплекс из современных методов ML / использует их некорректно.  
1	Студент использует несколько современных методов ML, но часть из них использована некорректно.  
2	Студент корректно использует несколько современных методов ML, приводит сравнение их результатов, но это не адекватно решаемой задаче (например, использует слишком большие и сложные модели для задач, которые того не требуют).  
3	Студент корректно использует комплекс из современных методов ML, приводит сравнение их результатов, и это адекватно решаемой задаче.  
**Валидация, постановка экспериментов: воспроизводимость модели (модель работает на новых данных) и кода в целом**  
0	Студент некорректно производит эксперименты, есть явные ошибки в используемых схемах валидации.  
1	Студент использует неустойчивые схемы валидации, воспроизводимость экспериментов под вопросом.  
2	Студент корректно ставит эксперименты, что гарантирует воспроизводимость решения.  
**Возможность реализации предложенного решения в продакшене**   
0	Предложенное решение не может быть реализовано в продакшене из-за низкого качества / неспособности решить поставленную задачу.  
1	Предложенное решение не может быть реализовано в продакшене в явном виде из-за его вычислительной сложности.  
2	Предложенное решение может быть с минимальными изменениями реализовано в продакшене.  
3	Предложенное решение уже реализовано студентом в продакшене.    
**Использование дополнительных инструментов, не рассматриваемых в курсе (дополнительные баллы)**  
0	Студент использует только рассмотренные в курсе инструменты/подходы.  
1	Студент использует какие-то инструменты, которые не были рассмотрены в курсе, но делает это формально.  
2	Студент активно использует инструменты, которые не были рассмотрены в курсе, и это улучшает качество его решения.  

**3. Оформление кода**

**Правильность, понятность, читабельность, документация к коду**

3	Есть недочёты в оформлении кода / документации к нему, но в целом код верен и понятен.  
5	Код верный, соответствует стандартам оформления, имеет документацию.
